```python
import pandas as pd                       # Importing packages
import matplotlib.pyplot as plt
import seaborn as sns
```

### *1. Read data in python environment* 

```python
cars_data= 'C:\SRUTHY B\cars_data.csv'    # Reading CSV file in Py environment
df = pd.read_csv(cars_data)
df
```

```python
df.info()            # Extracting informations from data
```

### *2. Null values in the given dataset*

```python
# Finding null values in the given set

df.isnull().sum()
```

no null values in the given dataset

### *3. Plot a bar graph of male vs female buyers participated in the sales*

```python
sns.set(font_scale=1.75)
df['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7,7),rot=0)
plt.xlabel("Gender",labelpad=14)
plt.ylabel("Count of Buyers",labelpad=14)
plt.title("Female vs Male Buyers participated in sales");
```

### *4. Top 5 cars based on their sales price*

```python
# Sorting data based on sales price
x=df.sort_values(by=['Sale Price'],ascending=False).head()

# Filtering data
x.loc[: ,['Make','Model','Sale Price']]
```

### *5. Least 5 cars based on their Resell price*

```python
# Sorting data values based on resell price
y=df.sort_values(by=['Resell Price'],ascending=True).head()

# Filtering data
y.loc[: ,['Make','Model','Resell Price']]
```
